```python
def create_lag_features(df, n):
    for i in range(1,n+1):
        df[f'lag_{i}'] = df['pollution'].shift(i)
    return df
```

That too when we have not used any hyperparameter tuning tool now lets use optuna and lets see how far we can push it.

```python
def objective(trial):
    param = {
        'objective': 'reg:squarederror',  # Regression task with squared error
        'eval_metric': 'rmse',            # Root Mean Squared Error (RMSE)
        'max_depth': trial.suggest_int('max_depth', 3, 10),  # Maximum depth of trees
        'learning_rate': trial.suggest_float('learning_rate', 1e-5, 1e-1, log=True),  # Learning rate
        'n_estimators': trial.suggest_int('n_estimators', 50, 500),  # Number of trees
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),  # Fraction of samples used for fitting
        'alpha': trial.suggest_float('alpha', 1e-5, 1e-1, log=True),  # L1 regularization term
        'lambda': trial.suggest_float('lambda', 1e-5, 1e-1, log=True),  # L2 regularization term
    }

    kf = KFold(n_splits=5, shuffle=True, random_state=42)

    model = xgbr(**param)
    scores = cross_val_score(model, x_train, y_train, cv=kf, scoring='r2')

    return np.mean(scores)
```

```python
study = optuna.create_study(direction='maximize')
```

```python
study.optimize(objective, n_trials=77, n_jobs=-1)
```

```python
study.best_params
```
